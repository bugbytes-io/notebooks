An important point from a memory-consumption view: **slicing creates a COPY of the elements in the original list**

```python
a[2:5]

# a[10_000:250_000] # create a copy of all 240,000 sliced items.
```

`itertools.islice()` addresses this issue. It creates an iterator over the items in the provided iterable

`a = [4,2,1,6,3,9,9,6]`

```python
islicer = itertools.islice(a, 0, 4)

# islicer
# for element in islicer:
#   print(element)

islicer = itertools.islice(a, 0, 5, 2)
list(islicer)
```

`islice()` can also slice iterators - not just sequences.

Normal slicing only works with sequences.

```python
gen = (i**2 for i in range(10))
gen
for num in gen:
  print(num)

# gen[1:4]  # will fail

slicer = itertools.islice(gen, 1, 4)
list(slicer)
```

```python
# get every second element squared from list "a"

squared = [num**2 for num in itertools.islice(a, 0, len(a), 2)]

for original, sq in zip(a[::2], squared):
  print(original, sq)
```

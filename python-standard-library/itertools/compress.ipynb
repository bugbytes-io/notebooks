### itertools.compress(iterable, mask)

The compress function takes an iterable and a mask, and returns only those elements in the iterable for which the associated mask value is truthy.

For example:



```
people = ['Larry David', 'Jeff Greene']
mask = [True, False]
list(itertools.compress(people, mask)) # ['Larry David']
```



````python
import itertools

languages = ['Python', 'JavaScript', 'Haskell', 'C++', 'Java', 'C#', 'Golang']
skill_mask = [True, True, False, False, True, False, True]

for language in itertools.compress(languages, skill_mask):
  print(f"I know {language}")
````

````python
for language in itertools.compress(languages, [not x for x in skill_mask]):
  print(f"I don't know {language}")
````

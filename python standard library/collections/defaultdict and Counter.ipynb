Often in Python, we want to count the number of occurrences in a data structure.

```python
people = ["John", "Dave", "Martin", "John", "Brad", "John", "Dave"]

# manually
counts = {}
for person in people:
  if person in counts:
    counts[person] += 1
  else:
    counts[person] = 1

counts
```

```python
# introduce defaultdict
from collections import defaultdict

counts = defaultdict(int)
for person in people:
  counts[person] += 1

counts
# issubclass(defaultdict, dict)
```

```python

from collections import Counter

counts = Counter(people)

counts
# issubclass(Counter, dict)

counts.most_common(2)
counts.subtract(['John', 'Brad'])
counts
```

```python
# applying to a tuple
Counter((1,2,3,4,2,3,6,8,3,4,6,4,5,5,5,5,5,5,5))

# apply to a set (1 per element)
Counter({1,2,3,4,2,3,6,8,3,4,6,4,5,5,5,5,5,5,5})
```
